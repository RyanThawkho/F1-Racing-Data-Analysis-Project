```python
#import libraries
import pandas as pd
import os
```

```python
#read the datasets
drivers_df = pd.read_csv("/content/drivers.csv", na_values=r'\N')
circuits_df = pd.read_csv("/content/circuits.csv", na_values=r'\N')
races_df = pd.read_csv("/content/races.csv", na_values=r'\N')
results_df = pd.read_csv("/content/results.csv", na_values=r'\N')
constructors_df = pd.read_csv("/content/constructors.csv", na_values=r'\N')
constructor_results_df = pd.read_csv("/content/constructor_results.csv", na_values=r'\N')
pit_stops_df = pd.read_csv("/content/pit_stops.csv", na_values=r'\N')
```

```python
drivers_df.head(5)
```

```python
#delete any duplicates
circuits_df = circuits_df.drop_duplicates()
drivers_df = drivers_df.drop_duplicates()
races_df = races_df.drop_duplicates()
results_df = results_df.drop_duplicates()
constructors_df = constructors_df.drop_duplicates()
constructor_results_df = constructor_results_df.drop_duplicates()
pit_stops_df = pit_stops_df.drop_duplicates()
```

```python
#drop unnecessary columns
drivers_df = drivers_df.drop(["number", "code", "url"], axis=1)
races_df = races_df.drop(races_df.columns[7:],axis=1)
constructors_df = constructors_df.drop(["url", "constructorRef"], axis=1)
```

```python
#rename columns for the drivers dataset
drivers_df = drivers_df.rename(columns={"dob": "date_of_birth",
                                        "forename": "first_name",
                                        "surname": "last_name"})


#rename columns for the circuits dataset
circuits_df = circuits_df.rename(columns={"location": "city",
                                          "lat": "Latitude",
                                          "lng": "Longitude",
                                          "alt": "Altitude"})

#rename columns for the results dataset
results_df = results_df.rename(columns={"position": "final_position",
                                        "rank": "fastest_lap_rank"})
```

```python
try:
  # Create an empty folder
  folder_name = "/content/clean_data"
  os.makedirs(folder_name, exist_ok=True)

except OSError as e:
  print(f"Error creating folder: {e}")

print(f"Folder '{folder_name}' created successfully.")
```

```python
#save datasets to a "clean_data" folder
try:
  drivers_df.to_csv("/content/clean_data/drivers.csv", index=False, encoding="utf-8-sig")
  circuits_df.to_csv("/content/clean_data/circuits.csv", index=False, encoding="utf-8-sig")
  races_df.to_csv("/content/clean_data/races.csv", index=False, encoding="utf-8-sig")
  results_df.to_csv("/content/clean_data/results.csv", index=False, encoding="utf-8-sig")
  constructors_df.to_csv("/content/clean_data/constructors.csv", index=False, encoding="utf-8-sig")
  constructor_results_df.to_csv("/content/clean_data/constructor_results.csv", index=False, encoding="utf-8-sig")
  pit_stops_df.to_csv("/content/clean_data/pit_stops.csv", index=False, encoding="utf-8-sig")

except OSError as e:
  print(f"Error creating folder: {e}")

print(f"Datasets copied successfully.")
```
